```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn import metrics
from sklearn.model_selection import train_test_split
from pandas import Series
data=pd.read_csv('../Data/weatherHistory.csv')
#Parameters precip type formatted date ,apparent temp, loud cover are useless
#Hence Drop them
final_data=data.drop( columns=['Formatted Date', 'Precip Type', 'Loud Cover','Apparent Temperature (C)'] )


#Summary is ambiguous hence removing for time being
#final_data=final_data.drop(columns=['Summary'])
y=[]
summary_val=[]
print(type(data['Summary']))
y_values=['Foggy','foggy','Overcast','overcast','cloudy','Cloudy','clear','Clear','rain','Rain','windy','Windy','breezy','Breezy','drizzle','Drizzle','Dry','dry']
classes_dict={'FOGGY':1,'OVERCAST':2,'CLOUDY':3,'CLEAR':4,'RAIN':5,'WINDY':6,'BREEZY':7,'DRIZZLE':8,'DRY':9}
for index, row in data.iterrows():
    test=0

    for i in y_values:
        if(not row['Daily Summary'].find(i)==-1):
            y.append(i.upper())
            break
    for i in y_values:
        if(not row['Summary'].find(i)==-1):
            summary_val.append(classes_dict[i.upper()])            
            break
final_data=final_data.drop(columns=['Summary'])
final_data['Summary'] = Series(summary_val, index=final_data.index)
#print((final_data['Summary']))

#daily summary is y for us
data_result=y
final_data=final_data.drop(columns=['Daily Summary'])
#print(type(data_result))
```

```python
len(data_result)
```

```python
from sklearn.model_selection import train_test_split

#Train Splitter

x_train, x_test, y_train, y_test = train_test_split(final_data,data_result, test_size=0.35, random_state=0)
print(y_train)
```

```python
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
import sys

# Function Name:sklearn.preprocessing.StandardScaler, 
# Input: X : array-like, shape [n_samples, n_features]         
# Output:
# Logic:Perform standardization by centering and scaling
# Example Call: scaler.transform(x_test)


scaler = StandardScaler()
scaler.fit(x_train)

x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)

print((x_test))
```

```python
from sklearn.linear_model import LogisticRegression as LR
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import average_precision_score
model = LR(multi_class="multinomial",solver="saga")

model.fit(x_train, y_train)


predictions = model.predict(x_test)


print(model.score(x_test, y_test))
print(metrics.f1_score(y_test,predictions,average='micro'))
print(metrics.confusion_matrix(y_test, predictions))

n_classes=['FOGGY','OVERCAST','CLOUDY','CLEAR','RAIN','WINDY','BREZZY','DRIZZLE','DRY']


```

```python
import joblib

   
filename_model = 'final_model.joblib'
filename_scaler='final_scaler.joblib'
joblib.dump(predictions, filename_model)
joblib.dump(scaler,filename_scaler)
```
